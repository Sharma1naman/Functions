**Question 1.** What is the difference between a function and a method in Python?
 - a function in Python is a block of reusable code that performs a specific task. Functions are defined using the def keyword followed by a name and parentheses that may include parameters. Functions help in breaking a program into smaller, manageable parts and can be called from anywhere in the code once defined. They are independent and can exist without being tied to any particular object.

 On the other hand, a method is similar to a function but is associated with an object (i.e., it belongs to a class). Methods are called on objects and can operate on data (attributes) that belongs to that class. For instance, in object-oriented programming examples, string operations like .upper() or .find() are methods because they act on string objects. Methods are implicitly passed the instance (usually named self) they are called on, while regular functions do not have this behavior.

```python
#Example of a function
def add_numbers(a, b):
    return a + b

result = add_numbers(5, 10)
print(result)
```

```python
#Example of a method
text = "hello world"
print(text.upper())
```

**Question 4.** What is the purpose of the return statement in a Python function?

**Question 5.**  What are iterators in Python and how do they differ from iterables?
 - In Python, an iterable is any object that you can loop over (use in a for loop). Examples of iterables are lists, tuples, strings, and dictionaries. An iterable object must implement the iter() method, which returns an iterator.
On the other hand, an iterator is an object that represents a stream of data; it remembers its current position while iterating. An iterator must implement two methods: iter() (which returns the iterator object itself) and next() (which returns the next element and moves the position forward).

  The main difference:

  **Iterable:** You can get an iterator from it (something you can loop over).

  **Iterator:** Actually produces the next value when you ask for it (using next()).



```python
#Example showing Iterable vs Iterator
# A list is an iterable
my_list = [1, 2, 3]
for item in my_list:
    print(item)

# Getting an iterator from the iterable
my_iterator = iter(my_list)
print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))
```

**Question 6.** Explain the concept of generators in Python and how they are defined.
  - In Python, a generator is a special type of iterator that is used to create an iterable sequence of values, but instead of returning all the values at once, it generates them one at a time using the yield keyword.
  Generators are memory-efficient because they do not store all values in memory — they produce values on demand.

  In Python, a generator is defined by creating a function that uses the yield keyword instead of return.
  When a function contains yield, it becomes a generator function.
  Each time the generator’s next() method is called (or you use next()), the function runs up to the next yield statement, returns the yielded value, and pauses its state until the next call. This is very different from a normal function that ends once return is called.

  The yield keyword allows the function to produce a series of values over time instead of computing them all at once and sending them back as a list.



```python
#Example of defining a generator
def countdown():
    yield 3
    yield 2
    yield 1
    yield "Go!"

# Create generator object
gen = countdown()

# Fetch values one by one
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

**Question 7.** What are the advantages of using generators over regular functions?

- The advantages of using generators over regular functions in Python are:

    - **Memory Efficiency:**
        Generators do not store all the values in memory. They generate one value at a time when needed, which is especially useful when working with large datasets. Regular functions (that return lists) build the entire list in memory at once, consuming more memory.

    - **Lazy Evaluation:**
     Generators produce items on-the-fly (only when requested). This means they are faster for large sequences where you don't need all the results immediately.

    - **Convenient for Infinite Sequences:**
     Generators can model infinite streams of data easily (like reading a huge file line-by-line), whereas regular functions would try to create the entire infinite list, causing the program to crash.

    - **Simpler Code for Iterators:**
     Writing generator functions with yield is simpler and cleaner than writing custom iterator classes with iter() and next() methods.


```python
#Example of Regular Function
def get_numbers_list():
    return [1, 2, 3]

# Call the function and print the result
numbers = get_numbers_list()
print(numbers)
```

```python
#Example of Generator Function
def get_numbers_generator():
    yield 1
    yield 2
    yield 3

# Create generator object
gen = get_numbers_generator()

# Get values using next()
print(next(gen))
print(next(gen))
print(next(gen))
```

**Question 8.** What is a lambda function in Python and when is it typically used?

 - A lambda function in Python is a small, anonymous function defined using the lambda keyword instead of the regular def statement. It can take any number of arguments but can only contain one expression. The result of that expression is automatically returned. Lambda functions are useful for short, throwaway functions — especially when you want to pass a simple function as an argument to functions like map(), filter(), reduce(), or sorted().

  Lambda functions are typically used when defining a function for a short period, such as sorting a list based on custom keys, or applying quick transformations to data.



```python
#Example of Regular function
def square(x):
    return x * x

print(square(5))
```

```python
#Example of equivalent Lambda Function
square_lambda = lambda x: x * x

print(square_lambda(5))
```

```python
#Typical use case — inside map()
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, numbers))
print(squared)
```

**Question 9.**  Explain the purpose and usage of the `map()` function in Python.

  - The map() function is used to apply a given function to every item in an iterable (like a list or tuple) and return a new map object (which is an iterator). It is commonly used when you want to perform the same operation on each element of a sequence — such as squaring numbers, converting strings to uppercase, or formatting data.


  

```python
# Example 1: Using map() with a regular function
def square(x):
    return x * x

numbers = [1, 2, 3, 4]
result = map(square, numbers)

print(list(result))
```

```python
# Example 2: Using map() with a lambda function
numbers = [1, 2, 3, 4]
result = map(lambda x: x ** 2, numbers)

print(list(result))
```

```python
# Example 3: Convert strings to uppercase
names = ['alice', 'bob', 'charlie']
result = map(lambda name: name.upper(), names)

print(list(result))
```

**Question 10.** What is the difference between map(), reduce(), and filter() functions in Python?

 - The main difference between map(), filter(), and reduce() lies in what they do with the data.

   - map() is used to **transform** each item in an iterable by applying a function to it — it returns a new iterable of the same length.

   - filter() is used to **select** items from an iterable based on a condition — it returns only the elements that satisfy that condition.

   - reduce(), from the functools module, is used to **combine** all items in an iterable into a single cumulative value (like sum or product) — it returns one final result, not a list.

  So in short:

   - map() = transform each item

   - filter() = choose some items

   - reduce() = boil down to one item



```python
# Example
from functools import reduce

numbers = [1, 2, 3, 4, 5, 6]

# map: square each number
squared = list(map(lambda x: x**2, numbers))
print("Map:", squared)

# filter: keep only even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers))
print("Filter:", evens)

# reduce: multiply all numbers
product = reduce(lambda x, y: x * y, numbers)
print("Reduce:", product)
```

**Question 1.**  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.



```python
def sum_of_evens(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

# Example usage:
my_list = [1, 2, 3, 4, 5, 6]
result = sum_of_evens(my_list)
print("Sum of even numbers:", result)
```

**Question 4.** Write a Python function that checks if a given number is prime or not from 1 to 200.



```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

# Example usage:
for number in range(1, 201):
    if is_prime(number):
        print(number)
```

**Question 5.** Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.




```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a

# Example usage:
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

**Question 6.** Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp

# Example usage:
for value in powers_of_two(5):
    print(value)
```

**Question 10.** Create a Python program that uses `filter()` to remove all the vowels from a given string.



```python
def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, input_str))

# Example usage:
text = "Hello, Python is amazing!"
result = remove_vowels(text)
print(result)
```
